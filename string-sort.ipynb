```python
import string
import random
import time
import matplotlib.pyplot as plt

# lengths = [i for i in range(0, 1000000, 1000)]
lengths = [10 ** i for i in range(1, 8)]
memo_times = []
sort_times = []
for i in lengths:
    original_string = ''.join(random.choice(string.ascii_uppercase) for _ in range(i))
    start = time.time()
    memo = [0] * 26
    for c in original_string:
        memo[ord(c) - ord('A')] += 1

    reconstructed = ''
    for i in range(26):
        reconstructed += chr(i + ord('A')) * memo[i]

    end = time.time()
    memo_times.append(end - start)

    start = time.time()
    reconstructed = ''.join(sorted(original_string))
    end = time.time()
    sort_times.append(end - start)
print(memo_times)
print(sort_times)
```

```python
plt.plot(lengths, memo_times, label='memo')
plt.plot(lengths, sort_times, label='sort')
plt.xscale('log')
plt.yscale('log')
plt.legend()
```
